```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data

```python
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
uniform_data = np.random.rand(10,12)
```

## Figure aesthetics

```python
# Create one figure and one subplot
f, ax = plt.subplots(figsize=(3,4))
```

```python
# Reset the Seaborn default style
sns.set()
```

```python
# Set style
sns.set_style('whitegrid')
```

```python
# See axes style
sns.axes_style('whitegrid')
```

```python
# Set context
sns.set_context('notebook', font_scale=1.2) 
```

```python
# Set color palette
sns.set_palette('husl')
```

## Plotting

```python
# Plot contrastive subsets of the data
sns.FacetGrid(titanic, col='survived', row='sex').map(plt.hist, 'age')
```

```python
# Plot categorical pointplot
sns.factorplot(x='pclass', y='survived', hue='sex', data=titanic)
```

```python
# Plot regression model fits across a FacetGrid
sns.lmplot(x='sepal_width', y='sepal_length', hue='species', data=iris)
```

```python
# PairGrid
sns.PairGrid(iris).map(plt.scatter)
```

```python
# Pairplot
sns.pairplot(iris)
```

```python
# Jointplot
sns.jointplot('sepal_length', 
              'sepal_width',
               data=iris,
               kind='kde')
```

```python
# Scatterplot
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width')
```

```python
# Stripplot
sns.stripplot(x='species', 
              y='petal_length',
              data=iris)
```

```python
# Swarmplot
sns.swarmplot(x='species',
              y='petal_length',
              data=iris)
```

```python
# Barplot
sns.barplot(x='sex',
            y='survived',
            hue='class',
            data=titanic)
```

```python
# Histplot
sns.histplot(data=titanic, 
             x='sex')
```

```python
# Countplot
sns.countplot(x='deck',
              data=titanic)
```

```python
# Pointplot
sns.pointplot(x='class',
              y='survived',
              hue='sex',
              data=titanic,
              markers=["^","o"],
              linestyles=["-","--"])
```

```python
# Boxplot
sns.boxplot(data=iris, orient='h')
```

```python
# Violinplot
sns.violinplot(x='age',
               y='sex',
               hue='survived',
               data=titanic)
```

```python
# Regplot
sns.regplot(x='sepal_width',
            y='sepal_length',
            data=iris)
```

```python
# Distplot
sns.displot(data=titanic, x='sex')
```

```python
# Heatmap
sns.heatmap(uniform_data, vmin=0, vmax=1)
```

## Customizations
